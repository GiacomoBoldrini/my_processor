```python
import time, random

def split_chunks(l, n):
    """ 
       Splits list l into n chunks with approximately equals sum of values
       see  http://stackoverflow.com/questions/6855394/splitting-list-in-chunks-of-balanced-weight
    """
    result = [[] for i in range(n)]
    sums   = {i:0 for i in range(n)}
    c = 0
    for e in l:
        for i in sums:
            if c == sums[i]:
                result[i].append(e)
                break
        sums[i] += e
        c = min(sums.values())    
    return result



MIN_VALUE = 1
MAX_VALUE = 10
ITEMS     = 1000
CHUNKS    = 260

l =[random.randint(MIN_VALUE, MAX_VALUE ) for i in range(ITEMS)]

t = time.time()

r = split_chunks(l, CHUNKS)
print(r)

print(ITEMS, CHUNKS, time.time() - t)
```

```python
import htcondor
def sort_dict(d, key_index=0, **kwargs):
    return dict(sorted(d.items(), key=lambda k: k[key_index], **kwargs))
def get_slot(slot):
    slot_number  = int(slot.split('@')[0][len('slot'):])
    slot_address = slot.split('@')[-1]
    return slot_address, slot_number
def get_summary_hosts(hosts):
    hosts_summary = {}
    for host in hosts:
        hosts_summary[host] = {}
        for slot in hosts[host]:
            user = hosts[host][slot]
            nslots_user = hosts_summary[host].get(user, 0)
            nslots_user += 1
            hosts_summary[host][user] = nslots_user
        #hosts_summary[host] = dict(sorted(hosts_summary[host].items(), key= lambda k: k[1], reverse=True))
        hosts_summary[host] = sort_dict(hosts_summary[host], 1, reverse=True)
    return hosts_summary
import json


coll = htcondor.Collector()

# Get all hosts and slots
projection = ["Machine", "State", "Name", "SlotID", "Activity", "MyAddress"]
slots_info = coll.query(htcondor.AdTypes.Startd, projection=projection)
hosts = {}
for slot in slots_info:
    host = slot.get('Name')
    state = slot.get('State')
    if host:
        hostname, slot = get_slot(host)
        #print(state)
        host_dict = hosts.get(hostname, {})
        if state == 'Unclaimed':
            host_dict[slot] = 'free'
        elif state == 'Claimed':
            host_dict[slot] = 'user'
        else:
            host_dict[slot] = 'unknown state ' + state
        hosts[hostname] = host_dict


# Get a schedd for each user (will only use the first one)
projection = ["Name", "MyAddress"]
all_submitters_query = coll.query(htcondor.AdTypes.Submitter, projection=projection)
schedds = []
for submitter in all_submitters_query:
    schedds.append(htcondor.Schedd(submitter))
projection = ["Owner", "User", "ExitStatus", "Cmd", "ClusterId", "ProcId",
                      "GlobalJobId", "JobStatus", "RemoteSlotID", "RemoteHost"]
jobs_from_submitter = list(schedds[0].query(projection=projection))
nslots_busy = 0
users = {}
for job in jobs_from_submitter:
    host = job.get('RemoteHost')
    user = job.get('User')
    if host:
        slot_number = host.split('@')[0]
        slot_address = host.split('@')[-1]
        hostname, slot = get_slot(host)
        hosts[hostname][slot] = user
        nslots_busy+=1
        user_dict = users.get(user, {})
        user_host = user_dict.get(hostname, 0)
        user_host += 1
        user_dict[hostname] = user_host
        users[user] = user_dict

for user in users:
    for host in users[user]:
        tot = len(hosts[host])
        users[user][host] = f"{users[user][host]}/{tot}"
    users[user] = sort_dict(users[user], 0)

users = sort_dict(users, 0)
print('Busy slots', nslots_busy)

for host in hosts:
    hosts[host] = sort_dict(hosts[host], 0)
hosts = sort_dict(hosts, 0)
print('\n\n')
print('Hosts')
print(json.dumps(get_summary_hosts(hosts), indent=2))
print('\n\n')
print('Users')
print(json.dumps(users, indent=2))

```
